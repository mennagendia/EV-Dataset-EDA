```python
EV_df['Efficiency'] = EV_df['Efficiency'].replace({'Wh/km':''}, regex=True).astype(int) #Efficiency -> int
EV_df['Range'] = EV_df['Range'].replace({'km':''}, regex=True).astype(int) # range -> int
EV_df['Acceleration'] = EV_df['Acceleration'].replace({'sec':''}, regex=True).astype(float) #Acceleration -> float
EV_df['TopSpeed'] = EV_df['TopSpeed'].replace({'km/h':''}, regex=True).astype(int) # top speed -> int
EV_df['PriceinUK'] = EV_df['PriceinUK'].replace({'\£': '', ',': ''}, regex=True).astype(float) # price in uk -> float
EV_df['PriceinGermany'] = EV_df['PriceinGermany'].replace({'\€': '', ',': ''}, regex=True).astype(float) # price in germany -> float
```
